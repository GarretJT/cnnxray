## Script to move & split dataset

```python
import os
import shutil
import random

base_dir = '/content/drive/MyDrive/DeepLearning/UAS/Xray'
positive_dir = os.path.join(base_dir, 'DvXray_Positive_Samples')
negative_dir = os.path.join(base_dir, 'DvXray_Negative_Samples')
output_dir = '/content/drive/MyDrive/DeepLearning/UAS/Xray/organized_dataset'

# Define paths for training, validation, and testing
train_dir = os.path.join(output_dir, 'training')
val_dir = os.path.join(output_dir, 'validation')
test_dir = os.path.join(output_dir, 'testing')

# Create necessary directories
for directory in [train_dir, val_dir, test_dir]:
    os.makedirs(os.path.join(directory, 'positive'), exist_ok=True)
    os.makedirs(os.path.join(directory, 'negative'), exist_ok=True)

# Function to move _SD.png files to the appropriate directory
def move_sd_files(file_list, target_dir, source_dir):
    for file in file_list:
        if file.endswith('_SD.png'):
            source_img_path = os.path.join(source_dir, file)
            target_img_path = os.path.join(target_dir, 'positive' if source_dir == positive_dir else 'negative', file)

            try:
                if os.path.exists(source_img_path):
                    shutil.move(source_img_path, target_img_path)
                    print(f"Moved {file} to {os.path.join(target_dir, 'positive' if source_dir == positive_dir else 'negative')}")
                else:
                    print(f"Image file not found: {file}")
            except Exception as e:
                print(f"Error moving {file}: {str(e)}")
        else:
            print(f"Skipped non '_SD.png' file: {file}")

# Function to split dataset (70% train, 15% validation, 15% test)
def split_dataset(files):
    random.shuffle(files)
    total_files = len(files)
    train_files = files[:int(0.7 * total_files)]
    val_files = files[int(0.7 * total_files):int(0.85 * total_files)]
    test_files = files[int(0.85 * total_files):]
    return train_files, val_files, test_files

# Process positive samples ending with _SD.png
positive_files = [f for f in os.listdir(positive_dir) if f.endswith('_SD.png')]
train_pos, val_pos, test_pos = split_dataset(positive_files)
move_sd_files(train_pos, train_dir, positive_dir)
move_sd_files(val_pos, val_dir, positive_dir)
move_sd_files(test_pos, test_dir, positive_dir)

# Process negative samples ending with _SD.png
negative_files = [f for f in os.listdir(negative_dir) if f.endswith('_SD.png')]
train_neg, val_neg, test_neg = split_dataset(negative_files)
move_sd_files(train_neg, train_dir, negative_dir)
move_sd_files(val_neg, val_dir, negative_dir)
move_sd_files(test_neg, test_dir, negative_dir)

print("Dataset organization completed.")
```

## CNN

### Pre-processing

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import os

# inisialisasi ulang
base_dir = '/content/drive/MyDrive/DeepLearning/UAS/Xray/organized_dataset'

train_dir = os.path.join(base_dir, 'training')
val_dir = os.path.join(base_dir, 'validation')
test_dir = os.path.join(base_dir, 'testing')

train_pos_dir = os.path.join(train_dir, 'positive')
train_neg_dir = os.path.join(train_dir, 'negative')
val_pos_dir = os.path.join(val_dir, 'positive')
val_neg_dir = os.path.join(val_dir, 'negative')
test_pos_dir = os.path.join(test_dir, 'positive')
test_neg_dir = os.path.join(test_dir, 'negative')
```

```python
train_datagen = ImageDataGenerator(rescale=1./255)
val_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(100, 100),
    batch_size=8,
    class_mode='binary'
)

val_generator = val_datagen.flow_from_directory(
    val_dir,
    target_size=(100, 100),
    batch_size=8,
    class_mode='binary'
)

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(100, 100),
    batch_size=8,
    class_mode='binary'
)
```

### Model

```python
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(100, 100, 3)),
    MaxPooling2D((2, 2)),
    Dropout(0.25),
    Flatten(),
    Dense(64, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])

model.summary()
```

### Train model

```python
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

history = model.fit(
    train_generator,
    steps_per_epoch=len(train_generator),
    epochs=10,
    validation_data=val_generator,
    validation_steps=len(val_generator)
)
```

```python
test_loss, test_accuracy = model.evaluate(test_generator, steps=len(test_generator))

print(f"Test Accuracy: {test_accuracy}")
```

### Load & Use saved model to predict

```python
import tensorflow as tf
from tensorflow.keras.preprocessing import image
import numpy as np

saved_model = tf.keras.models.load_model('/content/drive/MyDrive/DeepLearning/UAS/Xray/saved_model.keras')

def preprocess_image(img_path, target_size=(100, 100)):

    img = image.load_img(img_path, target_size=target_size)

    img_array = image.img_to_array(img)

    img_array = np.expand_dims(img_array, axis=0)
    # rubah citra agar mempunyai nilai 0-1
    img_array /= 255.0
    return img_array

def predict_image_class(model, img_path):
    img_array = preprocess_image(img_path)

    prediction = model.predict(img_array)
    print(prediction)

    class_label = 'positive' if prediction[0][0] > 0.5 else 'negative'
    return class_label

img_path = '/content/drive/MyDrive/DeepLearning/UAS/Xray/organized_dataset/testing/positive/P00017_OL.png'
class_label = predict_image_class(saved_model, img_path)
print(f'The predicted class for the image is: {class_label}')
```
